```python
#WAP to compute Central Tendency MeasuresL: Mean, Median, Mode.
import statistics

def central_tendency(numbers):
    print("Mean: ", statistics.mean(numbers))
    print("Median: ", statistics.median(numbers))
    print("Mode: ", statistics.mode(numbers))

numbers = [1, 2, 3, 4, 5, 5, 7, 8, 9, 10]
central_tendency(numbers)
```

```python
#WAP to compute Measures of Dispersion: Variance and Standard Deviation
def dispersion(numbers):
    print("Range: ", max(numbers) - min(numbers))
    print("Variance: ", statistics.variance(numbers))
    print("Standard Deviation: ", statistics.stdev(numbers))

numbers = [1, 2, 3, 4, 5, 5, 7, 8, 9, 10]
dispersion(numbers)
```

```python
#Scipy
import numpy as np
from scipy import stats

def central_tendency_dispersion(numbers):
    print("Mean: ", np.mean(numbers))
    print("Median: ", np.median(numbers))
    print("Mode: ", stats.mode(numbers)[0][0])
    print("Range: ", np.ptp(numbers))
    print("Variance: ", np.var(numbers))
    print("Standard Deviation: ", np.std(numbers))

numbers = [1, 2, 3, 4, 5, 5, 7, 8, 9, 10]
central_tendency_dispersion(numbers)
```

```python
#Pandas
import pandas as pd

def central_tendency_dispersion(numbers):
    series = pd.Series(numbers)
    print("Mean: ", series.mean())
    print("Median: ", series.median())
    print("Mode: ", series.mode().values[0])
    print("Range: ", series.max() - series.min())
    print("Variance: ", series.var())
    print("Standard Deviation: ", series.std())

numbers = [1, 2, 3, 4, 5, 5, 7, 8, 9, 10]
central_tendency_dispersion(numbers)
```

```python
#Linear Regression
import numpy as np

def simple_linear_regression(X, y):
    X = np.array(X).reshape(-1, 1)
    y = np.array(y).reshape(-1, 1)
    b = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    return b[0][0]

X = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
b = simple_linear_regression(X, y)
print("Slope (b): ",b)
```
